## ðŸ’° Section 1: ROI Calculation

### Comprehensive Financial Impact Analysis

```python
# Initialize Impact Calculator
calculator = ImpactCalculator()

# Calculate comprehensive ROI
roi_analysis = calculator.calculate_total_roi(enrolment_df, demographic_df, biometric_df)

print("=" * 80)
print(" " * 25 + "ðŸ’° ROI ANALYSIS SUMMARY")
print("=" * 80)

# Time Savings
print("\nðŸ“Š TIME SAVINGS:")
print("-" * 80)
ts = roi_analysis['time_savings']
print(f"  Manual Processing Hours:     {ts['manual_processing_hours']:>20,.0f} hrs")
print(f"  Automated Processing Hours:  {ts['automated_processing_hours']:>20,.0f} hrs")
print(f"  Hours Saved:                 {ts['hours_saved']:>20,.0f} hrs ({ts['time_savings_percentage']:.1f}%)")
print(f"  FTE Saved:                   {ts['fte_saved']:>20.2f} FTEs")
print(f"  Cost Savings:                {ts['cost_savings_inr']:>20,.0f} INR")

# Quality Improvement
print("\nâœ… QUALITY IMPROVEMENT:")
print("-" * 80)
qi = roi_analysis['quality_improvement']
print(f"  Errors Prevented:            {qi['errors_prevented']:>20,.0f}")
print(f"  Error Reduction:             {qi['error_reduction_percentage']:>19.1f}%")
print(f"  Error Cost Savings:          {qi['error_cost_savings_inr']:>20,.0f} INR")

# Decision Impact
print("\nðŸŽ¯ DECISION MAKING IMPACT:")
print("-" * 80)
di = roi_analysis['decision_impact']
print(f"  Annual Decisions:            {di['annual_decisions']:>20,}")
print(f"  Decision Hours Saved:        {di['decision_hours_saved']:>20,.0f} hrs")
print(f"  Decision Cost Savings:       {di['decision_cost_savings_inr']:>20,.0f} INR")
print(f"  Quality Improvement:         {di['decision_quality_improvement_percentage']:>19.0f}%")

# Costs
print("\nðŸ’µ COSTS:")
print("-" * 80)
costs = roi_analysis['costs']
print(f"  Implementation Cost:         {costs['implementation_cost_inr']:>20,.0f} INR (One-time)")
print(f"  Annual Maintenance:          {costs['annual_maintenance_inr']:>20,.0f} INR/year")
print(f"  Total 3-Year Cost:           {costs['total_3yr_cost_inr']:>20,.0f} INR")

# Benefits
print("\nðŸ’š BENEFITS:")
print("-" * 80)
benefits = roi_analysis['benefits']
print(f"  Annual Benefits:             {benefits['annual_benefits_inr']:>20,.0f} INR/year")
print(f"  Net Benefit (Year 1):        {benefits['net_benefit_year1_inr']:>20,.0f} INR")
print(f"  Net Benefit (Year 2+):       {benefits['net_benefit_year2_onwards_inr']:>20,.0f} INR/year")
print(f"  Total 3-Year Benefits:       {benefits['total_3yr_benefits_inr']:>20,.0f} INR")

# ROI
print("\nðŸ† ROI METRICS:")
print("-" * 80)
roi = roi_analysis['roi']
print(f"  Payback Period:              {roi['payback_period_months']:>19.1f} months")
print(f"  3-Year ROI:                  {roi['roi_3yr_percentage']:>19.1f}%")
print(f"  Break-Even Achieved:         {str(roi['break_even_achieved']):>20}")
print("=" * 80)
```

```python
# Visualize ROI breakdown
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Cost vs Benefits (3-year)
ax1 = axes[0, 0]
categories = ['Total Costs\n(3 Years)', 'Total Benefits\n(3 Years)', 'Net Profit\n(3 Years)']
values = [
    costs['total_3yr_cost_inr'],
    benefits['total_3yr_benefits_inr'],
    benefits['total_3yr_benefits_inr'] - costs['total_3yr_cost_inr']
]
colors = ['#EF476F', '#06D6A0', '#3A86FF']
bars1 = ax1.bar(categories, values, color=colors, alpha=0.8, edgecolor='black', linewidth=2)
for bar, val in zip(bars1, values):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
            f'â‚¹{val/1e6:.1f}M',
            ha='center', va='bottom', fontweight='bold', fontsize=11)
ax1.set_ylabel('Amount (INR)', fontweight='bold', fontsize=12)
ax1.set_title('3-Year Financial Overview', fontweight='bold', fontsize=14, pad=15)
ax1.grid(True, alpha=0.3, axis='y')
ax1.ticklabel_format(style='plain', axis='y')

# 2. Annual Benefits Breakdown
ax2 = axes[0, 1]
benefit_components = {
    'Time Savings': ts['cost_savings_inr'],
    'Quality\nImprovement': qi['error_cost_savings_inr'],
    'Decision\nImpact': di['decision_cost_savings_inr']
}
wedges, texts, autotexts = ax2.pie(
    benefit_components.values(),
    labels=benefit_components.keys(),
    autopct='%1.1f%%',
    startangle=90,
    colors=['#3A86FF', '#FB5607', '#06D6A0']
)
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(11)
ax2.set_title('Annual Benefits Breakdown', fontweight='bold', fontsize=14, pad=15)

# 3. Payback Period Timeline
ax3 = axes[1, 0]
months = list(range(0, 37))
cumulative = []
for m in months:
    if m == 0:
        cumulative.append(-costs['implementation_cost_inr'])
    else:
        monthly_benefit = (benefits['annual_benefits_inr'] - costs['annual_maintenance_inr']) / 12
        cumulative.append(cumulative[-1] + monthly_benefit)

ax3.plot(months, cumulative, linewidth=3, color='#3A86FF', marker='o', markersize=4)
ax3.axhline(y=0, color='red', linestyle='--', linewidth=2, label='Break-Even Point', alpha=0.7)
ax3.fill_between(months, 0, cumulative, where=[c >= 0 for c in cumulative], 
                 color='green', alpha=0.2, label='Profit Zone')
ax3.fill_between(months, 0, cumulative, where=[c < 0 for c in cumulative], 
                 color='red', alpha=0.2, label='Investment Zone')
ax3.set_xlabel('Months', fontweight='bold', fontsize=12)
ax3.set_ylabel('Cumulative Net Benefit (INR)', fontweight='bold', fontsize=12)
ax3.set_title('Payback Period Timeline (36 Months)', fontweight='bold', fontsize=14, pad=15)
ax3.grid(True, alpha=0.3)
ax3.legend(loc='lower right', fontsize=10)
ax3.ticklabel_format(style='plain', axis='y')

# 4. ROI Comparison
ax4 = axes[1, 1]
roi_metrics = {
    'Year 1': ((benefits['net_benefit_year1_inr']) / (costs['implementation_cost_inr'] + costs['annual_maintenance_inr'])) * 100,
    'Year 2': ((benefits['net_benefit_year2_onwards_inr'] * 2 + benefits['net_benefit_year1_inr']) / (costs['implementation_cost_inr'] + costs['annual_maintenance_inr'] * 2)) * 100,
    'Year 3': roi['roi_3yr_percentage']
}
bars4 = ax4.bar(roi_metrics.keys(), roi_metrics.values(), 
               color=['#FFD60A', '#FB5607', '#06D6A0'], alpha=0.8, edgecolor='black', linewidth=2)
for bar, val in zip(bars4, roi_metrics.values()):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height,
            f'{val:.1f}%',
            ha='center', va='bottom', fontweight='bold', fontsize=12)
ax4.set_ylabel('ROI (%)', fontweight='bold', fontsize=12)
ax4.set_title('ROI Progression (Year-over-Year)', fontweight='bold', fontsize=14, pad=15)
ax4.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('../outputs/figures/roi_comprehensive_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nâœ“ ROI visualizations created")
```
