### Innovation #1: Aadhaar Health Index (AHI)

**Purpose**: Measure overall system health with 4 components:
- **Coverage (30%)**: State/district penetration
- **Update Velocity (25%)**: Daily update rate
- **Data Quality (25%)**: Completeness + validity
- **Accessibility (20%)**: Geographic equity + temporal availability

**Scale**: 0-100 (Higher = Better)  
**Rating**: EXCELLENT (80+), GOOD (60-80), FAIR (40-60), NEEDS IMPROVEMENT (<40)

```python
# Initialize Aadhaar Health Index calculator
ahi_calculator = AadhaarHealthIndex()

# Calculate comprehensive index
ahi_result = ahi_calculator.calculate_composite_index(
    enrolment_df, demographic_df, biometric_df
)

print("ðŸ¥ AADHAAR HEALTH INDEX (AHI)")
print("=" * 70)
print(f"Overall AHI Score: {ahi_result['aadhaar_health_index']:.2f}/100")
print(f"Rating: {ahi_result['rating']}")
print("\nðŸ“Š Component Breakdown:")
print("-" * 70)
for component, score in ahi_result['components'].items():
    print(f"  {component.replace('_', ' ').title():.<50} {score:.2f}")
print("=" * 70)

# Visualize components
fig, ax = plt.subplots(figsize=(10, 6))
components = list(ahi_result['components'].keys())
scores = list(ahi_result['components'].values())

bars = ax.barh(components, scores, color=['#3A86FF', '#FB5607', '#8338EC', '#06D6A0'])
ax.set_xlabel('Score (0-100)', fontweight='bold', fontsize=12)
ax.set_title('Aadhaar Health Index - Component Scores', fontweight='bold', fontsize=14, pad=20)
ax.set_xlim(0, 100)

# Add value labels
for i, (bar, score) in enumerate(zip(bars, scores)):
    ax.text(score + 2, i, f'{score:.1f}', va='center', fontweight='bold')

plt.tight_layout()
plt.savefig('../outputs/figures/ahi_component_breakdown.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nâœ“ AHI analysis complete")
```
